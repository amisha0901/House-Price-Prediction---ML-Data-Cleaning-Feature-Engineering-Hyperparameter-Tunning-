```python
valid = ['Alley','BsmtQual','BsmtCond','BsmtExposure','BsmtFinType1','BsmtFinType2','FireplaceQu','GarageType', 'GarageFinish','GarageQual','GarageCond','PoolQC','Fence','MiscFeature']
d1 = {}
for col in df_test.columns:
    if col in valid:
        d1[col] = ['']
    else:
        d1[col] = ['','NA']
d1
```

```python
df_train = pd.read_csv('../../datasets/house_train.csv',keep_default_na=False, na_values=d1) 
df_test = pd.read_csv('../../datasets/house_test.csv',keep_default_na=False, na_values=d1)     #'keep_default_na=False' shows missing value
```

```python
df_train.info()
```

# Outlier Handling

```python
df_train = df_train[df_train['SalePrice']<340000].reset_index(drop=True)
```

```python
df_train.info()
```

```python
cols_numeric = list(df_test.select_dtypes(exclude='object').columns)
cols_object = list(df_test.select_dtypes(include = 'object').columns)
```

```python
 # jar df_train getla tar saleprice column yenar
 # we take df_test because we dont want y col(saleprice) because we cant fill missing value of saleprice.
```

# Missing values

```python
from sklearn.impute import SimpleImputer
```

```python
imputer_numeric = SimpleImputer(strategy='median')
imputer_numeric.fit(df_train[cols_numeric]) # calculate median of every column
df_train[cols_numeric] = imputer_numeric.transform(df_train[cols_numeric]) # fill missing values

imputer_object = SimpleImputer(strategy='most_frequent')
imputer_object.fit(df_train[cols_object]) # calculate mode ...
df_train[cols_object] = imputer_object.transform(df_train[cols_object])
```

```python
df_train.isna().sum().sum()
```

```python
df_test.isna().sum().sum()
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df_train[cols_numeric])
df_train[cols_numeric] = scaler.transform(df_train[cols_numeric])
```

# Encoding

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False,handle_unknown='ignore')
encoder.fit(df_train[cols_object]) # finds categories in object cols
```

```python
cols_encoded = encoder.get_feature_names_out()               # It has only columns name
len(cols_encoded)
```

```python
df_train[cols_encoded] = encoder.transform(df_train[cols_object])
```

```python
df_train.shape
```

# Model BUilding

```python
cols_final = cols_numeric+list(cols_encoded)
X_train = df_train[cols_final]
y_train = df_train['SalePrice']
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

# Make Predictions

```python
# in .fit we cant use test data thats why we use .transform

df_test[cols_numeric] = imputer_numeric.transform(df_test[cols_numeric])
df_test[cols_object] = imputer_object.transform(df_test[cols_object])

df_test[cols_numeric] = scaler.transform(df_test[cols_numeric])
df_test[cols_encoded] = encoder.transform(df_test[cols_object])

X_test = df_test[cols_final]
yp = model.predict(X_test)
```
